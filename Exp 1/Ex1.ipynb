```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/prave/Downloads/Car Sales.xlsx - car_data.csv/Car Sales.xlsx - car_data.csv")

df.columns = df.columns.str.strip().str.lower() 
print("Column Names:", df.columns) 


df['date'] = pd.to_datetime(df['date'])  

sales_per_day = df.groupby('date').size().reset_index(name='sales')

sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_per_day, x='date', y='sales', label='Car Sales', color='blue')


plt.xlabel('Date')
plt.ylabel('Number of Cars Sold')
plt.title('Car Sales Over Time')
plt.xticks(rotation=45)
plt.show()


```
